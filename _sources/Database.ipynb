# Tugas - Database Power BI

Aplikasi yang dibutuhkan :
1. PostgreSql
2. Power Bi
3. XAMPP
4. Elephantsql.com


 Langkah-langkah untuk mengumpulkan data dari beberapa database
 

**1. Membuat database di server cloud di postgres**

1. Membuat instance baru pada akun elephantsql.com dengan nama pendatab yang nantinya
digunakan sebagai server cloud.
2. Menghubungkan server dari elephantsql.com ke postgres dengan cara :
a. klik kanan pada server > klik menu register > klik server
b. beri nama server
c. isi bagian connection sesuai dengan server yang ada pada instance pendatab di akun
elephantsql.com
host name/ address diisi dengan server
maintenance database dan username diisi dengan user dan default database
password diisi sesuai dengan yang tertera pada detail instance elephantsql
kemudian klik tombol save, dan server pendatab berhasil dibuat.
3. Membuat tabel data iris
membuat database berdasarkan nama user daan default database dari instance akun
elephantsql yaitu : fhkviitd
klik kanan pada menu tabel dan pilih menu query tool
kemudian tuliskan create table dari data yang ingin di inputkan
lalu run query yang sudah dibuat dan refresh pada menu table untuk memunculkan
tabel data iris yang sudah dibuat

**2. Database menggunakan localhost di postgresql**
1. Buat server bernama localpendatab, kemudian mengisi bagian connection lalu klik save
2. Membuat tabel data iris
a. membuat database bernama databaseiris
b. membuat tabel iris dengan cara yang sama dengan pembuatan tabel irirs server cloud
c. tabel irirs berhasil dibuat

**3. Power BI**

**data iris yang terdapat pada database localhost **
1. export table iris di postgresql menjadi file csv
2. input file csv dengan cara klik get data > text/csv pada Power BI
3. kemudian klik transform data
4. kemudian kita bisa me remove column/rows sesuai data apa saja yang ingin kita tampilkan
analisisnya
5. kemudian load data
6. kemudian akan muncul data iris di bagian kanan layar, centang semuanya
7. kemudian di bagian visualization pilih model chart yang ingin digunakan untuk menampilkan grafik data iris

**data iris yang terdapat pada database cloud postgres **
1. klik getdata > more > cari postgres database > klik postgres database > klik connect
2. masukkan nama server sesuai yang ada pada elephantsql.com dan nma databasenya
disesuaikan dengan postgres > klik ok
3. kemudian isikan usernam, password dan server yang sama dengan yang ada di elephantsql
lalu klik connect > load > selesai.

**data iris yang terdapat pada database MySQL **
1. klik menu get data > more > cari mysql database > klik mysql database > connect
2. kemudian isis nama server dengan locahost diikuti dengan port mysql pada XAMPP dan
database sesuai dengan nama database di phpmyadmin
3. kemudian klik load
4. jika sudah tampil di power bi, pilih tampilan visualnya sebagai tabel dan tabel data irirs akan
terlihat.
data iris csv ke phpmyadmin
1. start apache dan mysql pada XAMPP
2. buka phpmyadmin > buat database bernama pendatairis
3. import csv ke phpmyadmin
4. kemudian data tabel akan tampil jika berhasi
